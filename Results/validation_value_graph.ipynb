```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
results = pd.read_csv('validation_log.csv')
```

### Validation Log for MovieLens-10M

```python
plt.vlines(x=100, ymin=0, ymax=1)
plt.plot(results['10m_validation_ndcg100_bivae_ract'], color='red', label='BiVAE-RaCT')
plt.plot(results['10m_validation_ndcg100_bivae'], color='black', label='BiVAE')
plt.xlim(xmin=80, xmax=120)
plt.ylim(ymin=0.45, ymax=0.53)
plt.xlabel('# Epoch of Actor')
plt.ylabel('Validation NDCG@100')
plt.title('ML-10M')
plt.legend()
plt.show()
```
